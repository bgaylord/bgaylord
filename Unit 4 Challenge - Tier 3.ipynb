```python
#https://stackoverflow.com/questions/55776571/how-to-split-a-date-column-into-separate-day-month-year-column-in-pandas
properties["day"] = properties['date'].map(lambda x: x.day)
properties["month"] = properties['date'].map(lambda x: x.month)
properties["year"] = properties['date'].map(lambda x: x.year)
print(properties)
```

```python
properties_by_year = properties.groupby(['variable','year']).mean().reset_index()
properties_by_year.head()
```

```python
#extract a subset for a given year
properties_1995 = properties_by_year[properties_by_year.year==1995].set_index('variable')
properties_1995.head()
```

```python
properties_2021 = properties_by_year[properties_by_year.year==2021].set_index('variable')
properties_2021.head()
```
